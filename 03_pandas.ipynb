## Pandas

Pakiet Pandas udostępnia dwa obiekty służące do wygodnego przechowywania danych i ułatwiających manipulowanie danymi:
* **Series** to obiekt reprezentujący jednowymiarową tablicę danych
* **DataFrame** to dwu-wymiarowa tabela danych (zbiór obiektów series)

Pandas WWW: <https://pandas.pydata.org/>  
Pandas Tutorial: <https://pandas.pydata.org/docs/getting_started/intro_tutorials/index.html>

## Series i DataFrame

## Series

Seria jest obiektem reprezentującym jednowymiarową tablicę danych o indeksowanych elementach. Seria może być utworzona z dowolnej kolekcji liczb: krotki, listy, tablicy `numpy` a nawet ze słownika. Obiekt `Series` udostępnia wiele metod pozwalających operować na danych w serii, zawiera też  metody do wizualizowania danych. Dodatkowo, wiele funkcji z pakietu `numpy` może być użytych na tym obiekcie, jeżeli tylko seria zawiera liczby.

```python
import pandas as pd

s = pd.Series( [3.1, 2.4, -1.7, 0.2, -2.9, 4.5] )   # seria utworzona z listy liczb
print(s)
```

```python
s = pd.Series([3.1, 2.4, -1.7, 0.2, -2.9, 4.5]) 

print('Values=', s.values, type(s.values))     # wartości serii (obiekt ndarray z NumPy)
print('Index=', s.index)                       # indeksy serii 
print('Pierwszy element:', s[0])               # indeksowanie elementów
```

W odróżnieniu do tablic indeksami nie muszą być liczby całkowite. 

```python
import numpy as np

s = pd.Series(np.random.randn(5), index=['a', 'b', 'c', 'd', 'e'], name='Zmienna')
s
```

```python
print('Values=', s.values)   # wartości, to tablica numpy
print('Index=', s.index)     # indeksy
```

```python
print(s['a'])     # indeksowanie za pomocą napisów
print(s[0])       # indeksowanie liczbami całkowitymi nadal działa
```

Obiekt Series w wielu sytuacjach zachowuje się podobnie do tablic NumPy.  
W podobny sposób indeksuje się elementy serii, z tą różnicą, że przekroje również dotyczą indeksów.

```python
s = pd.Series([1.2, 2.5, -2.2, 3.1, -0.8], index = ['pn','wt','śr','czw','pt'], name='Zmienna')
s
```

```python
s[1:3]          # przekrój serii (elementy od 1 do 2)
```

```python
s['wt':'czw']    # elementy od wt do czw (zauważ, że ostatni element 'czw' też jest włączony)
```

Gdy seria zawiera wartości numeryczne to można wykonywać na niej operacje arytmetyczne i funkcje NumPy

```python
s = s +  4
s
```

```python
s = np.sqrt(s)
s
```

## DataFrame

Obiekt `DataFrame` reprezentuje 2-wymiarową tabelę danych, która posiada indeksowane kolumny oraz wiersze.  
Każda kolumna to obiekt Series. Kolumna reprezentuje zmienną i może zawierać elementy różnego typu (liczby, wartości kategoryczne, napisy, serie czasowe, itd.). 

<div>

</div>

Tablica DataFrame posiada wiele przydatnych metod ułatwiających operowanie danymi, wyznaczanie statystyk i sporządzenie wykresów.

Zobacz: <https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.html>

Tworzenie tablic DataFrame (np. z tablicy numpy)

```python
df = pd.DataFrame(np.random.rand(4,3), columns=['zmienna 1', 'zmienna 2', 'zmienna 3'])
df
```

Własność ``values`` zwraca tablicę w reprezentacji Numpy

```python
print('Values=', df.values)
print('Typ', type(df.values))
```

```python
print('Index=',  df.index)      
print('Columns:', df.columns)
```

Kolumny (zmienne) indeksowane są za pomocą nazw zmiennych 

```python
df['zmienna 2']   # wynikiem jest seria
```

Kolumna tabeli `DataFrame` jest obiektem `Series`

```python
type(df['zmienna 2'])
```

Dodawanie nowych kolumn w tabeli przypomina dodawanie nowych elementów do słownika

```python
df['rok'] = 2021                              # dodanie kolumny ze stałą wartością
df
```

```python
df['x1'] = np.random.rand(4)     # dodawanie sekwencji wartosci
df['liczby'] = ['raz', 'dwa', 'trzy', 'cztery']
df
```

Tworzenie nowej zmiennej jako funkcji wynikowej innych zmiennych

```python
df['zm1+zm2'] = df['zmienna 1'] + df['zmienna 2']
df
```

Wybór kilku zmiennych (indeksem jest lista zmiennych)

```python
df[ ['rok', 'zmienna 1'] ]
```

Wynik jest typu ``DataFrame``, otrzymujemy nowy widok tabeli

```python
type(df[ ['zmienna 1', 'rok'] ]) 
```

Kolumna (seria) dostępna jest też poprzez własność obiektu ``DataFrame`` o tej samej nazwie co nazwa zmiennej

```python
df.rok           # to samo co df['rok']
```

```python
df.liczby[2:]
```

### Indeksowanie tablic DataFrame

* atrybut `iloc[]` - indeksowanie za pomocą __liczb całkowitych__ (lub macierzy boolowskich) identycznie jak macierze `numpy`
* atrybut `loc[]` - indeksowanie za pomocą etykiet (indeks, nazwy zmiennych) lub macierzy boolowskich

```python
display(df)
```

```python
df.iloc[2]    # wiersz 3, wynikiem jest seria z elementów wiersza, elementy mogą być różnego typu
```

```python
df.iloc[2 ,1]    # 3 wiersz, 2 kolumna, 
```

Własność ``iloc``: indeksowanie tablicy jak w Numpy

```python
display(df)
df.iloc[:, 2]   # trzecia kolumna tabeli -> wynikiem jest seria
```

```python
df.iloc[:2, 1:3]     # przekrój danych (wiersze 0,1,2 i kolumny 1,2) -> wynik DataFrame
```

Własność ``loc``: indeksowanie nazwami zmiennych i kluczami indeksu

```python
df.loc[2]    # gdy indeksy są liczbami to działa to analigicznie jak iloc[]
```

```python
df.loc[:2]    # zauważ, że loc[] uwzględnia ostatni element zakresu
```

```python
df.loc[:2, 'rok':'x1']   # drugi wymiar to zmienne, można indeksować po nazwach
```

```python
df.loc[:2, ['rok', 'x1']]
```

Filtrowanie za pomocą wyrażeń logicznych

```python
df[ df['x1'] < 0.6 ]
```

```python
df[ (df.x1 < 0.6) | (df.liczby == "trzy") ]
```

### Grupowanie danych

```python
print('Średnie wartości w grupie mężczyzn:')
tips[tips.sex == 'Male'][['tip', 'total_bill']].mean()
```

Operacja grupowania `groupby()` pozwala w wygodny sposób wykonać operacje na grupach danych (np. wyznaczyć wartość średnią w grupach).  

Funkcje agregujące: ``mean(), min(), max(), count(), first()``, itd.

```python
tips_mean = tips[ ['total_bill', 'tip', 'sex'] ].groupby(['sex']).mean()    # średnie wartości napiwków (tips) względem płci 
tips_mean                                                                   # wynik jest typu DataFrame
```

```python
tips_mean = tips[ ['tip', 'sex'] ].groupby(['sex']).mean()
tips_mean['tip'].plot(kind='bar', title='Mean tip');
```

Wyznaczmy liczebność próbek w grupie mężczyzn i kobiet (zmienna grupująca ``sex``)

```python
tips_count = tips.groupby(['sex']).count()
display(tips_count)
tips_count.plot.pie(y='tip', autopct='%1.0f%%', legend=False,);
```
